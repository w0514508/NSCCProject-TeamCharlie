# Atlantic Data Append
This lab is appending files for all province together

```python
import pandas as pd
from pathlib import Path
```

```python
NB = pd.read_csv("../../data/clean/Yashaswi/AtlanticData/NB_30Y_clean.csv", dtype={"Climate ID": "string"}, low_memory=False)
NS = pd.read_csv("../../data/clean/Yashaswi/AtlanticData/NS_30Y_clean.csv", dtype={"Climate ID": "string"}, low_memory=False)
NL = pd.read_csv("../../data/clean/Yashaswi/AtlanticData/NL_30Y_clean.csv", dtype={"Climate ID": "string"}, low_memory=False)
PE = pd.read_csv("../../data/clean/Yashaswi/AtlanticData/PE_30Y_clean.csv", dtype={"Climate ID": "string"}, low_memory=False)
```

```python

folder_path = Path("../../data/clean/Yashaswi/AtlanticData/")
csv_files = folder_path.glob("*.csv")

Atlantic_all = pd.concat(
    (
        pd.read_csv(
            file,
            dtype={"Climate ID": "string"},  # ✅ force correct dtype
            low_memory=False                  # ✅ avoid chunk guessing
        )
        for file in csv_files
    ),
    ignore_index=True
)
```

```python
Atlantic_all.info()
```

```python

str_cols = [
    "Station Name",
    "Climate ID",
    "Date/Time",
    "Spd of Max Gust (km/h)"
]

Atlantic_all[str_cols] = Atlantic_all[str_cols].astype("string")
```

```python
Atlantic_all.info()
```

```python
Atlantic_all.to_csv("Atlantic_all_30Y_clean.csv", index=False)
```
