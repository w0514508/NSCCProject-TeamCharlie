# Station data Append
This lab is appending files for all years together

```python

import pandas as pd
from pathlib import Path

folder_path = Path("../../data/raw/Yashaswi/NS Data/")   # adjust if needed
csv_files = folder_path.glob("*.csv")

NS_all = pd.concat(
    (pd.read_csv(file, encoding="latin-1") for file in csv_files),
    ignore_index=True
)

```

```python
NS = NS_all.drop(
    columns=[
        'Month', 'Day',
        'Data Quality',
        'Max Temp Flag', 'Min Temp Flag', 'Mean Temp Flag',
        'Heat Deg Days Flag', 'Cool Deg Days Flag', 'Total Rain Flag',
        'Total Snow Flag', 'Total Precip Flag', 'Snow on Grnd Flag',
       'Spd of Max Gust Flag',  'Dir of Max Gust Flag']
)
```

```python
NS.info()
```

```python
NS['Date/Time'] = pd.to_datetime(NS['Date/Time'])
```

```python
NS['Station Name'] = NS['Station Name'].astype('string')
```

```python
NS['Climate ID'] = NS['Climate ID'].astype('string')
```
