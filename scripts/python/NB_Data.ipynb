# Station data Append
This lab is appending files for all years together

```python

import pandas as pd
from pathlib import Path

folder_path = Path("../../data/raw/Yashaswi/NB data/")   # adjust if needed
csv_files = folder_path.glob("*.csv")

NB_all = pd.concat(
    (pd.read_csv(file, encoding="latin-1") for file in csv_files),
    ignore_index=True
)

```

```python
NB = NB_all.drop(
    columns=[
        'Month', 'Day',
        'Data Quality',
        'Max Temp Flag', 'Min Temp Flag', 'Mean Temp Flag',
        'Heat Deg Days Flag', 'Cool Deg Days Flag', 'Total Rain Flag',
        'Total Snow Flag', 'Total Precip Flag', 'Snow on Grnd Flag',
       'Spd of Max Gust Flag',  'Dir of Max Gust Flag']
)
```

```python
NB.info()
```

```python
NB['Date/Time'] = pd.to_datetime(NB['Date/Time'])
```

```python
NB['Station Name'] = NB['Station Name'].astype('string')
```

```python
NB['Climate ID'] = NB['Climate ID'].astype('string')
```

```python
NB.to_csv("NB_30Y_clean.csv", index=False)
```
