# Station data Append
This lab is appending files for all years together

```python

import pandas as pd
from pathlib import Path

folder_path = Path("../../../data/raw/Yashaswi/NL Data/")   # adjust if needed
csv_files = folder_path.glob("*.csv")

NL_all = pd.concat(
    (pd.read_csv(file, encoding="latin-1") for file in csv_files),
    ignore_index=True
)

```

```python
NL = NL_all.drop(
    columns=[
        'Month', 'Day',
        'Data Quality',
        'Max Temp Flag', 'Min Temp Flag', 'Mean Temp Flag',
        'Heat Deg Days Flag', 'Cool Deg Days Flag', 'Total Rain Flag',
        'Total Snow Flag', 'Total Precip Flag', 'Snow on Grnd Flag',
       'Spd of Max Gust Flag',  'Dir of Max Gust Flag']
)
```

```python
NL.info()
```

```python
NL['Date/Time'] = pd.to_datetime(NL['Date/Time'])
```

```python
NL['Station Name'] = NL['Station Name'].astype('string')
```

```python
NL['Climate ID'] = NL['Climate ID'].astype('string')
```

```python
NL['Climate ID'].unique()
```

```python
NL.info()
```

```python
NL.to_csv("NL_30Y_clean.csv", index=False)
```
