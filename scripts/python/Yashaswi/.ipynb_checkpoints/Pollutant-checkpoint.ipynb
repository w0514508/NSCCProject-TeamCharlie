```python
import pandas as pd
from pathlib import Path

# Folder containing input files
DATA_DIR = Path("../../../data/raw/Yashaswi/AQI/")

# Pollutant → averaging period to extract
AVERAGING_MAP = {
    "PM2.5": "24HR",
    "PM10": "24HR",
    "O3": "8HR",
    "CO": "1HR",
    "NO": "1HR",
    "NO2": "1HR",
    "NOX": "1HR",
    "SO2": "1HR",
}

def read_pollutant_file(file_path):
    """
    Read a CSV/XLSX file, extract the correct averaging table,
    and append the year parsed from the filename.
    """
    year = int(file_path.name[:4])  # first 4 characters
    name_lower = file_path.stem.lower()

    pollutant = next(
        (p for p in AVERAGING_MAP if p.lower() in name_lower),
        None
    )

    if pollutant is None:
        return None

    avg_id = AVERAGING_MAP[pollutant]

    if file_path.suffix == ".csv":
        df = pd.read_csv(file_path)
        df = df[df["ID"] == avg_id]

    elif file_path.suffix in (".xls", ".xlsx"):
        
        df = pd.read_excel(
            file_path,
            sheet_name=avg_id
        )


    else:
        return None

    df["year"] = year
    df["pollutant"] = pollutant

    return df


# ---- Batch load all files ----
frames = []
for file in DATA_DIR.glob("*.[cx][sl]*"):  # csv / xls / xlsx
    df = read_pollutant_file(file)
    if df is not None:
        frames.append(df)

final_df = pd.concat(frames, ignore_index=True)

# ---- Keep ONLY requested columns ----
COLUMNS_TO_KEEP = [
    "year",
    "pollutant",
    "NAPS ID",
    "City",
    "P/T",
    "Location",
    "Latitude",
    "Longitude",
    "Elevation",
    "ID",
    "Minimum",
    "Maximum",
    "Percent Completeness",
    "January",
    "February",
    "March",
    "April",
    "May",
    "June",
    "July",
    "August",
    "September",
    "October",
    "November",
    "December",
    "Standard Deviation",
]

final_df = final_df[[c for c in COLUMNS_TO_KEEP if c in final_df.columns]]

print(final_df.head())
```
