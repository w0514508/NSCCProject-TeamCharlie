```python

import pandas as pd
from pathlib import Path
```

```python
DATA_DIR = Path(r"../../../data/raw/Yashaswi/AQIPollutants/Allhourly/")  # update this
```

## STEP 2 – Reader for raw NAPS CSV

```python
def read_raw_naps_csv(path):
    df = pd.read_csv(
        path,
        skiprows=7,              # header row
        sep=",",                 # comma-separated
        engine="python",         # robust parser
        on_bad_lines="skip"      # skip malformed rows safely
    )

    # Clean bilingual column names
    df.columns = [c.split("//")[0].strip() for c in df.columns]
    return df
```

## STEP 3 – Read & append EVERYTHING automatically

```python
all_data = []

for file in DATA_DIR.glob("*.csv"):
    # Parse pollutant and year from filename
    name = file.stem            # e.g. "NO2_2021"
    pollutant, year = name.split("_")
    
    df = read_raw_naps_csv(file)
    
    # Add metadata columns
    df["Pollutant"] = pollutant
    df["Year"] = int(year)
    
    all_data.append(df)

raw_air_quality = pd.concat(all_data, ignore_index=True)
```
