```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ghg_province_clean.csv")

# Summary
print(df.describe())

# Trend per province
for province in df["Provinces"].unique():
    subset = df[df["Provinces"] == province]
    plt.plot(subset["Year"], subset["Emissions"], label=province)

plt.legend()
plt.title("GHG Emissions by Province Over Time")
plt.show()
```
