# Monthly Precipitation Data Cleaning – ECCC

This notebook performs data inspection, cleaning, and preparation of monthly precipitation data from Environment and Climate Change Canada (ECCC).

Scope of this notebook (Week 2):
- Monthly climate data
- Total Precipitation (mm)
- Nova Scotia weather stations
- Data wrangling and cleaning only (no analysis or modeling)

```python
import pandas as pd
import numpy as np
```

```python
# Load raw monthly climate data: Yarmouth station
file_path = "../Capstone/raw data/en_climate_monthly_NS_8205700_1941-2014_P1M_SYDNEY.csv"
df_raw = pd.read_csv(file_path)
```

```python
# Initial inspection of raw data
df_raw.head()
df_raw.info()
df_raw.describe(include="all")
```

The raw dataset contains multiple climate variables.
For the purposes of Week 2, only the Total Precipitation (mm) field is retained to demonstrate data cleaning and preparation steps.

```python
# Selecting relevant columns only
columns_to_keep = [
    "Year",
    "Month",
    "Total Precip (mm)"
]

df = df_raw[columns_to_keep].copy()
df.head()
```

```python
# Rename columns for consistency and ease of use
df.columns = [
    "year",
    "month",
    "total_precip_mm"
]
```

```python
# Data type cleaning
df["year"] = pd.to_numeric(df["year"], errors="coerce")
df["month"] = pd.to_numeric(df["month"], errors="coerce")
df["total_precip_mm"] = pd.to_numeric(df["total_precip_mm"], errors="coerce")

df.info()
```

```python
# Check missing values
df.isna().sum()
```

```python
# Remove records with missing precipitation or incomplete date information
df = df.dropna(subset=["year", "month", "total_precip_mm"])

# Convert year and month to integer after removing invalid values
df["year"] = df["year"].astype(int)
df["month"] = df["month"].astype(int)

df.info()
```

```python
# Check for duplicate records
df.duplicated().sum()
```

Records with missing precipitation values or incomplete date information were removed to produce a clean and consistent monthly dataset suitable for trend analysis. No duplicate values found.

```python
# Basic outlier inspection for precipitation
df["total_precip_mm"].describe()
```

```python
# Inspect extreme values using quantiles
df["total_precip_mm"].quantile([0.01, 0.99])
```

Potential outliers in monthly precipitation were inspected using summary statistics.
Extreme values were retained at this stage, as they may represent valid climatic events rather than data errors.
Outlier handling will be revisited during the analysis phase if required.

```python
# Create a proper date field for temporal consistency
df["date"] = pd.to_datetime(
    df["year"].astype(str) + "-" + df["month"].astype(str) + "-01"
)

df.head()
```

```python
# Final inspection of cleaned dataset
df.info()

# Save cleaned dataset for further analysis and visualization
output_path = "../Capstone/clean data/precipitation_monthly_clean_sydney.csv"
df.to_csv(output_path, index=False)

output_path
df.head()
```
