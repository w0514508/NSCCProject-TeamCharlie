## GHG Emissions in Canada

```python
#LOAD THE RAW DATA

import pandas as pd

df = pd.read_csv("GHG Provinces & Territories Raw Data.csv")

df.head()
df.info()
```

```python
#CHECK WHAT ALL THE COLUMNS TO DETERMINE USAGE
print(df.columns)
print("Shape of dataset:", df.shape)
```

```python
#KEEP ONLY WHAT IS NEEDED
df = df[["REF_DATE", "GEO", "VALUE", "Sector"]]
```

```python
print(df.columns)
```

```python
#RENAME COLUMNS (MAKE IT CLEAN)
df.columns = ["Year", "Provinces", "Emissions", "Sectors"]
```

```python
print(df.columns)
```

```python
#REMOVE NULLS
df = df.dropna()
```

```python
#FIX DATA TYPES
df["Year"] = df["Year"].astype(int)
df["Emissions"] = df["Emissions"].astype(float)
```

```python
#FILTER LAST 10 YEARS (PROJECT REQUIREMENT)
latest_year = df["Year"].max()
df = df[df["Year"] >= latest_year - 10]
```

```python
#REMOVE TOTAL / DUPLICATES
df = df[df["Provinces"] != "Canada"]
```

```python
#CREATE CLEAN DATASETS FOR PROVINCES (FOR MAIN CHARTS)

df_province = df.groupby(["Year", "Provinces"])["Emissions"].sum().reset_index()

df_province.to_csv("ghg_province_clean.csv", index=False)
```

```python
# CREATE CLEAN DATASETS FOR SECTOR DATA (NS ONLY)
df_sector_ns = df[df["Provinces"] == "Nova Scotia"]

df_sector_ns = df_sector_ns.groupby(["Year", "Sectors"])["Emissions"].sum().reset_index()

df_sector_ns.to_csv("ghg_sector_ns_clean.csv", index=False)
```
