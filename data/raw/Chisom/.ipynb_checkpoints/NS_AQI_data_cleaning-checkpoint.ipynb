```python
import pandas as pd
import numpy as np


df_aylesford = pd.read_csv(
    r"C:\Users\mickl\AQI\Nova_Scotia_level_Ozone_O3Aylesford.csv"
)
```

```python
df_aylesford.head()
df_aylesford.info()
```

```python
aylesford = df_aylesford.copy()
```

```python
aylesford = aylesford[["Date & Time", "Average"]]
```

```python
#  Rename columns
aylesford = aylesford.rename(
    columns={
        "Date & Time": "Date",
        "Average": "Concentration"
    }
)
aylesford.head()
```

```python
# Remove nulls
aylesford = aylesford.dropna()
```

```python
# Remove invalid (negative) concentrations

aylesford = aylesford[aylesford["Concentration"] >= 0]
```

```python
#Standardize Date (NO hour)
aylesford["Date"] = pd.to_datetime(
    aylesford["Date"], errors="coerce"
).dt.date
# Drop any failed conversions
aylesford = aylesford.dropna(subset=["Date"])
```

```python
#Add AQI columns
aylesford["Pollutant"] = "O3"
aylesford["Station"] = "Aylesford"
aylesford.info()
aylesford.head()
```

```python
aylesford.to_csv(
    r"C:\Users\mickl\AQI\Aylesford_O3_Staging.csv",
    index=False
)
```

```python
df_pictou = pd.read_csv(
    r"C:\Users\mickl\AQI\Nova_Scotia_level_Ozone_O3Pictou.csv"
)
```

```python
pictou = df_pictou.copy()
pictou = pictou[["Date & Time", "Average"]]

#Rename columns for AQI correctness
pictou = pictou.rename(
    columns={
        "Date & Time": "Date",
        "Average": "Concentration"
    }
)
```

```python
# Remove null values
pictou = pictou.dropna()
# Remove invalid (negative) concentrations
pictou = pictou[pictou["Concentration"] >= 0]
```

```python
#Standardize the Date column (no datetime / no hour)
pictou["Date"] = pd.to_datetime(
    pictou["Date"], errors="coerce"
).dt.date

#Remove bad parses
pictou = pictou.dropna(subset=["Date"])
```

```python
# Add AQI metadata columns
pictou["Pollutant"] = "O3"
pictou["Station"] = "Pictou"
pictou.info()
pictou.head()
```

```python
pictou.to_csv(
    r"C:\Users\mickl\AQI\Pictou_O3_Staging.csv",
    index=False
)
```

```python
df_sydney_no2 = pd.read_csv(
    r"C:\Users\mickl\AQI\Nova_Scotia_Nitrogen_Oxides_NO2Sydney.csv"
)
```

```python
#Creating a working copy
sydney_no2 = df_sydney_no2.copy()

# Select required columns
sydney_no2 = sydney_no2[["Date & Time", "Average"]]

# Rename columns for AQI correctness
sydney_no2 = sydney_no2.rename(
    columns={
        "Date & Time": "Date",
        "Average": "Concentration"
    }
)

# Remove null values
sydney_no2 = sydney_no2.dropna()

# Remove invalid (negative) concentrations
sydney_no2 = sydney_no2[sydney_no2["Concentration"] >= 0]
```

```python
# Standardize the Date column
sydney_no2["Date"] = pd.to_datetime(
    sydney_no2["Date"], errors="coerce"
).dt.date

# Drop failed date parses
sydney_no2 = sydney_no2.dropna(subset=["Date"])

# Add AQI metadata columns
sydney_no2["Pollutant"] = "NO2"
sydney_no2["Station"] = "Sydney"

sydney_no2.info()
sydney_no2.head()
```

```python
sydney_no2.to_csv(
    r"C:\Users\mickl\AQI\Sydney_NO2_Staging.csv",
    index=False
)
```

```python
df_sable_island_so2 = pd.read_csv(
    r"C:\Users\mickl\AQI\Nova_Scotia_Sulphur_Dioxide_SO2Sable_Island.csv"
)
# Create a working copy
sable_island_so2 = df_sable_island_so2.copy()
```

```python
# Select required columns
sable_island_so2 = sable_island_so2[["Date & Time", "Average"]]

# Rename columns for AQI correctness
sable_island_so2 = sable_island_so2.rename(
    columns={
        "Date & Time": "Date",
        "Average": "Concentration"
    }
)
```

```python
# Remove null values
sable_island_so2 = sable_island_so2.dropna()

# Remove invalid (negative) concentrations
sable_island_so2 = sable_island_so2[sable_island_so2["Concentration"] >= 0]

# Standardize the Date column (no datetime)
sable_island_so2["Date"] = pd.to_datetime(
    sable_island_so2["Date"], errors="coerce"
).dt.date

# Drop failed conversions
sable_island_so2 = sable_island_so2.dropna(subset=["Date"])

# Add AQI metadata columns
sable_island_so2["Pollutant"] = "SO2"
sable_island_so2["Station"] = "Sable Island"

sable_island_so2.info()
sable_island_so2.head()
```

```python
sable_island_so2.to_csv(
    r"C:\Users\mickl\AQI\Sable_Island_SO2_Staging.csv",
    index=False
)
```
