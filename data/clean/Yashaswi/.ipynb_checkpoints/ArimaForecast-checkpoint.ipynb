# Simple ARIMA for short-term temperature forecasting

## Prepare the time series (Atlantic Canada)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
import pandas as pd

df = pd.read_csv("../../clean/Yashaswi/Final_Clean/temp_clean_Daily.csv", 
                    dtype={"Climate ID": "string"},
                    low_memory=False
)
stations = pd.read_csv("../../clean/Yashaswi/Final_Clean/dim_station_clean.csv")

df["Date"] = pd.to_datetime(df["Date"])
df["Year"]  = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"]   = df["Date"].dt.day

df = df.rename(columns={"Climate ID": "station_id"})
stations = stations.rename(columns={"Climate ID": "station_id"})
df = df.merge(
    stations,
    on="station_id",
    how="left"
)
df.head()
```

```python
# Clean Province strings
df["Province"] = df["Province"].str.strip()
```

```python
atlantic_provinces = [
    "NOVA SCOTIA",
    "NEW BRUNSWICK",
    "NEWFOUNDLAND AND LABRADOR",
    "PRINCE EDWARD ISLAND"
]

df_atlantic = df[df["Province"].isin(atlantic_provinces)].copy()
```

```python
df_atlantic["Province"].value_counts()
```

```python
df.info()
```

```python
BASELINE_START = 1990
BASELINE_END   = 2026

df_baseline = df[
    (df["Year"] >= BASELINE_START) &
    (df["Year"] <= BASELINE_END)
]
df_baseline.head(3)
```

```python
baseline = (
    df_baseline
    .groupby(["station_id", "Month", "Day"])["Mean Temp (°C)"]
    .mean()
    .reset_index(name="Baseline Mean Temp (°C)")
)
```

```python

# Drop existing baseline/anomaly columns if they exist
df = df.drop(
    columns=["Baseline Mean Temp (°C)", "Temp Anomaly (°C)"],
    errors="ignore"
)

df = df.merge(
    baseline,
    on=["station_id", "Month", "Day"],
    how="left"
)
```

```python
df["Temp Anomaly (°C)"] = (
    df["Mean Temp (°C)"] - df["Baseline Mean Temp (°C)"]
)
```

```python
df["Temp Anomaly (°C)"].describe()
```

```python
atlantic_provinces = [
     "NOVA SCOTIA",
    "NEW BRUNSWICK",
    "NEWFOUNDLAND AND LABRADOR",
    "PRINCE EDWARD ISLAND"
]

df_atlantic = df[df["Province"].isin(atlantic_provinces)].copy()
df_atlantic["Province"].value_counts()
```

# Atlantic annual anomaly time series

```python
atlantic_annual_anomaly = (
    df_atlantic
    .groupby("Year")["Temp Anomaly (°C)"]
    .mean()
    .reset_index()
)
```

```python
atlantic_annual_anomaly.head(3)
```

```python
len(atlantic_annual_anomaly)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.plot(
    atlantic_annual_anomaly["Year"],
    atlantic_annual_anomaly["Temp Anomaly (°C)"]
)
plt.axhline(0, linestyle=":")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (°C)")
plt.title("Atlantic Canada Annual Mean Temperature Anomalies")
plt.tight_layout()
plt.show()
```

```python
ts = atlantic_annual_anomaly.set_index("Year")["Temp Anomaly (°C)"]
```

#  Step 5 — Prepare the time series for ARIMA
- ARIMA requires a 1‑D time series with a time index.

```python
ts = (
    atlantic_annual_anomaly
    .set_index("Year")["Temp Anomaly (°C)"]
    .sort_index()
)
```

```python


len(ts)
```

```python
ts.info()
```

## How to interpret this

Clear upward tendency → non‑stationary
Year‑to‑year variability → expected
✅ Differencing will be required → ARIMA(d=1)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.plot(ts.index, ts.values, marker="o")
plt.axhline(0, linestyle=":")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (°C)")
plt.title("Atlantic Canada Annual Temperature Anomalies")
plt.tight_layout()
plt.show()
```

## Step 7 — Fit a simple ARIMA model
Recommended climate‑safe starting model
ARIMA(1,1,1)
(robust, avoids overfitting, standard for short series)

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(ts, order=(1, 1, 1))
model_fit = model.fit()
```

```python
model_fit.summary()
```

## Step 8 — Short‑term forecast (5 years)

```python
forecast_years = 5

forecast = model_fit.get_forecast(steps=forecast_years)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()
```

## Step 9 — Plot observed + forecast

```python
plt.figure(figsize=(9,4))

# Observed
plt.plot(ts.index, ts.values, label="Observed", marker="o")

# Forecast
plt.plot(
    forecast_mean.index,
    forecast_mean.values,
    linestyle="--",
    marker="o",
    label="ARIMA forecast"
)

# Confidence interval
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    alpha=0.3,
    label="95% confidence interval"
)

plt.axhline(0, linestyle=":")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (°C)")
plt.title("Atlantic Canada: Short‑term Temperature Forecast (ARIMA)")
plt.legend()
plt.tight_layout()
plt.show()
```

### “Short‑term statistical forecasting was performed using an ARIMA(1,1,1) model applied to annual mean temperature anomalies. The forecast suggests persistence of recent positive anomaly conditions over the next several years, with substantial interannual variability. The ARIMA results complement, rather than replace, the linear trend analysis, which captures the longer‑term forced warming signal across Atlantic Canada.”

```python
plt.figure(figsize=(10,5))

# Observed
plt.plot(ts.index, ts.values, color="black", label="Observed")

# Forecast
plt.plot(
    forecast_mean.index,
    forecast_mean.values,
    linestyle="--",
    color="red",
    linewidth=2,
    label="Forecast"
)

# CI
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    color="red",
    alpha=0.2,
    label="95% CI"
)

# Forecast start line
plt.axvline(forecast_mean.index[0], color="gray", linestyle="--")

plt.axhline(0, linestyle=":", color="gray")

plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (°C)")
plt.title("Atlantic Canada Temperature Forecast (ARIMA)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")

best_aic = float("inf")
best_order = None

for p in range(4):
    for q in range(4):
        try:
            model = ARIMA(ts, order=(p, 0, q), trend="t")
            result = model.fit()
            
            if result.aic < best_aic:
                best_aic = result.aic
                best_order = (p, 0, q)
                
            print(f"(p,q)=({p},{q}) → AIC: {result.aic:.2f}")
        except:
            pass

print("\nBest model:", best_order, "AIC:", best_aic)
```

```python
from statsmodels.tsa.arima.model import ARIMA

p, d, q = best_order

model = ARIMA(ts, order=(p, d, q), trend="t")
model_fit = model.fit()
```

```python
model_fit.summary()
```

```python
forecast = model_fit.get_forecast(steps=5)

forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()
```

```python
# --- OLD MODEL ---
old_model = ARIMA(ts, order=(1,1,1))
old_fit = old_model.fit()

old_forecast = old_fit.get_forecast(steps=5)
forecast_mean_old = old_forecast.predicted_mean


# --- NEW MODEL ---
model = ARIMA(ts, order=(p,0,q), trend="t")
model_fit = model.fit()

forecast = model_fit.get_forecast(steps=5)
forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()


# --- PLOT ---
plt.figure(figsize=(10,5))

# Observed
plt.plot(ts.index, ts.values, color="black", label="Observed")

# Old
plt.plot(
    forecast_mean_old.index,
    forecast_mean_old.values,
    linestyle="--",
    color="blue",
    label="Old ARIMA (flat)"
)

# New
plt.plot(
    forecast_mean.index,
    forecast_mean.values,
    linestyle="--",
    color="red",
    label="ARIMA with trend"
)

# CI
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    color="red",
    alpha=0.2
)

plt.axvline(forecast_mean.index[0], color="gray", linestyle="--")

plt.legend()
plt.show()
```

##  (quick visualization — acceptable approximation)
👉 Convert yearly forecast into monthly points (smooth interpolation)
This does NOT add new information, but it lets you visualize finer detail.

```python
ts_monthly = df.groupby("Date")["Temp Anomaly (°C)"].mean()
ts_monthly = ts_monthly.resample("ME").mean()
```

## fit SARIMA (with seasonality!)

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(
    ts_monthly,
    order=(p,0,q),
    seasonal_order=(1,1,1,12),  # yearly seasonality
    trend="t"
)

model_fit = model.fit()
```

## Forecast monthly

```python
forecast = model_fit.get_forecast(steps=60)  # 5 years monthly

forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()
```

## Full visualization (historical + forecast)

```python
plt.figure(figsize=(12,5))

# Observed data
plt.plot(
    ts_monthly.index,
    ts_monthly.values,
    color="black",
    label="Observed"
)

# Forecast
plt.plot(
    forecast_mean.index,
    forecast_mean.values,
    color="red",
    linestyle="--",
    label="SARIMA Forecast"
)

# Confidence interval
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    color="red",
    alpha=0.2,
    label="95% CI"
)

# Forecast start marker
plt.axvline(forecast_mean.index[0], color="gray", linestyle="--")

plt.axhline(0, linestyle=":", color="gray")

plt.xlabel("Date")
plt.ylabel("Temperature Anomaly (°C)")
plt.title("Monthly Temperature Forecast (SARIMA)")
plt.legend()

plt.tight_layout()
plt.show()
```

##  Zoom into FUTURE ONLY (2026–2030)

```python
plt.figure(figsize=(12,5))

plt.plot(
    forecast_mean.index,
    forecast_mean.values,
    color="red",
    linestyle="--",
    label="Forecast"
)

plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    color="red",
    alpha=0.2
)

plt.xlim(
    pd.Timestamp("2026-01-01"),
    pd.Timestamp("2030-12-31")
)

plt.axhline(0, linestyle=":", color="gray")

plt.title("Monthly Forecast (Zoomed: 2026–2030)")
plt.xlabel("Date")
plt.ylabel("Temp Anomaly (°C)")
plt.legend()

plt.show()
```

## CLEAR seasonal pattern view

```python
plt.figure(figsize=(12,5))

# Last few years of observed
plt.plot(
    ts_monthly[-36:],  # last 3 years
    label="Recent observed",
    color="black"
)

# First part of forecast
plt.plot(
    forecast_mean[:24],  # first 2 years
    linestyle="--",
    color="red",
    label="Forecast"
)

plt.title("Seasonality Check (Observed vs Forecast)")
plt.legend()
plt.show()
```

## FULL SARIMA PIPELINE (monthly temperature)


###===============================
 1.IMPORT LIBRARIES
===============================


```python

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings

warnings.filterwarnings("ignore")
```


===============================
2. LOAD DATA
===============================


```python
df = pd.read_csv("../../clean/Yashaswi/Final_Clean/monthly_temperature_1995_2025.csv")
```

```python

# Convert Date column
df["Date"] = pd.to_datetime(df["Date"])
```

### 
===============================
3. PREPARE MONTHLY TIME SERIES
===============================
- If your data is already monthly, this ensures it's clean and sorted


```python
df.info()
```

```python
# Use mean temperature
ts_monthly = (
    df[["Date", "Monthly_Average_Mean_Temp_C"]]
    .dropna()
    .sort_values("Date")
    .set_index("Date")
    ["Monthly_Average_Mean_Temp_C"]
)
ts_monthly = ts_monthly.rolling(3).mean().dropna()
```

### STEP 4: Fit SARIMA

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(
    ts_monthly,
    order=(2, 0, 2),               # p, d, q
    seasonal_order=(1, 1, 1, 12),  # P, D, Q, s
    trend="t"
)
model_fit = model.fit()
```

## STEP 5: Check model

```python
print(model_fit.summary())
```

### STEP 6: Forecast (5 years)

```python
forecast = model_fit.get_forecast(steps=60)

forecast_mean = forecast.predicted_mean
forecast_ci = forecast.conf_int()
```

```python
last_date = ts_monthly.index[-1]
```

```python
future_dates = pd.date_range(
    start=last_date + pd.offsets.MonthEnd(1),
    periods=60,
    freq="ME"
)
```

```python
forecast_mean.index = future_dates
forecast_ci.index = future_dates
```

### STEP 7: Full visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

# Observed
plt.plot(
    ts_monthly.index,
    ts_monthly.values,
    color="black",
    label="Observed"
)

# Forecast
plt.plot(
    forecast_mean.index,
    forecast_mean.values,
    color="red",
    linestyle="--",
    linewidth=2,
    label="SARIMA Forecast"
)

# Confidence interval
plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    color="red",
    alpha=0.2
)

# Forecast start
plt.axvline(forecast_mean.index[0], color="gray", linestyle="--")

plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.title("Monthly Mean Temperature Forecast (SARIMA)")
plt.legend()

plt.tight_layout()
plt.show()
```

### STEP 8: Zoom into 2026–2030

```python
plt.figure(figsize=(12,5))

plt.plot(
    forecast_mean.index,
    forecast_mean.values,
    color="red",
    linestyle="--",
    label="Forecast"
)

plt.fill_between(
    forecast_ci.index,
    forecast_ci.iloc[:, 0],
    forecast_ci.iloc[:, 1],
    color="red",
    alpha=0.2
)

import pandas as pd
plt.xlim(pd.Timestamp("2026-01-01"), pd.Timestamp("2030-12-31"))

plt.axhline(0, linestyle=":", color="gray")

plt.title("Forecast (2026–2030)")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")

plt.legend()
plt.show()
```
